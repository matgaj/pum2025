```python
import numpy as np
from matplotlib import pyplot as plt 
#alternatywny zapis:
#import matplotlib.pyplot as plt
```

### Wykresy liniowe - funkcja `plot`

Jeżeli do funkcji plot podamy pojedynczy wektor danych, to zostanie przyjęte założenie, że jest to ciąg wartości y - na osi x pojawią się wtedy numery próbek od 0 do n-1, gdzie n to długość wektora (w Pythonie indeksowanie danych jest zawsze od 0, a nie od 1, jak w Matlabie).

```python
plt.plot([1, 2, 3, 4])
plt.ylabel('wartości próbki')
plt.xlabel('numer próbki')
plt.title('Przykładowe użycie funkcji plot - 1 wektor danych')
plt.grid(True) #włącz siatkę
plt.show()
```

Jeżeli podamy dwa wektory danych, to pierwszy będzie interpretowany jako współrzędne X, a drugi - współrzędne Y.

```python
plt.plot([1, 2, 3, 4], [4, 7, 3.4, 6], linewidth=4.0)
plt.ylabel('wartości próbki')
plt.xlabel('numer próbki')
plt.title('Przykładowe użycie funkcji plot - 2 wektory danych')
plt.xticks(np.arange(1, 5, step=1)) #określamy, z jakiego zakresu i z jakim krokiem mają być wyświetlone 
                                    #wartości na osi X
plt.grid(False) #wyłącz siatkę - opcja domyślna, można pominąć tę linijkę kodu, jeżeli nie chcemy siatki
plt.show()
```

### Wykresy punktowe - funkcje `plot` i `scatter`

Wykresy punktowe można rysować na dwa sposoby:
1. przy użyciu funkcji plot
2. przy użyciu funkcji scatter

```python
plt.plot([1, 2, 3, 4], [6, 7, 3.4, 1], 'go') #g oznacza kolor zielony, o - kształt markera
plt.ylabel('y')
plt.xlabel('x')
plt.title('Przykładowe użycie funkcji plot - wykres punktowy')
plt.show()
```

Uwaga: użycie poniższego kodu spowoduje, że uzyskamy wykres liniowy z dodatkowo zaznaczonymi punktami pomiarowymi, a nie wykres punktowy:

```python
plt.plot([1, 2, 3, 4], [6, 7, 3.4, 1], color='g', marker='o', linestyle='--') #zmieniamy linię na przerywaną
plt.ylabel('y')
plt.xlabel('x')
plt.title('Przykładowe użycie funkcji plot - wykres liniowy z zaznaczonymi punktami pomiarowymi')
plt.show()
```

```python
plt.scatter([1, 2, 3, 4], [6, 7, 3.4, 1], c='g')
plt.ylabel('y')
plt.xlabel('x')
plt.title('Przykładowe użycie funkcji scatter - liczby na osi X')
plt.show()
```

Obie funkcje mogą przyjmować zarówno dane liczbowe, jak i kategorialne (wyrażone słownie, stringiem) jako dane na osi X.

```python
values = [6, 7, 3.4, 1]
categories = ['a', 'b', 'c', 'd']
plt.figure(figsize=(15, 3)) #tworzymy figurę, na której będą narysowane wykresy i definiujemy jej wymiary:
                            #pierwszy oznacza wymiar w osi X - szerokość, a drugi w osi Y - wysokość
plt.subplot(121) #subplot działa tak samo jak w matlabie
plt.plot(categories, values, 'go')
plt.ylabel('wartości próbki')
plt.xlabel('kategoria')
plt.title('Przykładowe użycie funkcji plot - kategorie na osi X')
plt.legend(['Dane'], loc='upper left')
plt.subplot(122)
plt.scatter(categories, values, c='g')
plt.ylabel('wartości próbki')
plt.xlabel('kategoria')
plt.title('Przykładowe użycie funkcji scatter - kategorie na osi X')
plt.legend(['Dane'])
plt.show()
```

Użycie tych funkcji różni się w przypadku, gdy mamy więcej niż jedną serię danych i chcemy przedstawić je na jednym wykresie. Poniżej przykład dla 2 serii danych - jeżeli mają tę samą dziedzinę, to do funkcji plot wystarczy podać wektor wartości X oraz macierz wartości Y.

W przypadku funkcji scatter trzeba użyć jej tyle razy, ile serii chcemy wyświetlić i przypisać je do tego samego wykresu - oba wektory danych muszą mieć tę samą długość.

```python
values2D = np.array([[6, 3],[7,1], [3.4,11], [1, 8.2]])
```

```python
categories = ['a', 'b', 'c', 'd']
categories2D = [['a','a'], ['a','a'], ['a','a'], ['a','a']]
fig = plt.figure(figsize=(15, 3)) #tworzymy figurę, na której będą narysowane wykresy i definiujemy jej wymiary:
                            #pierwszy oznacza wymiar w osi X - szerokość, a drugi w osi Y - wysokość

#funkcja plot    
ax1 = fig.add_subplot(121)
ax1.plot(categories, values2D, marker='o', linestyle='None') #linestyle='None' usuwa linię z wykresu
ax1.set_ylabel('wartości próbki')
ax1.set_xlabel('kategoria')
ax1.set_title('Przykładowe użycie funkcji plot - dwie serie danych')
ax1.legend(['seria1', 'seria2'])

#funkcja scatter
ax2 = fig.add_subplot(122)
ax2.scatter(categories, values2D[:,0], c='g')
ax2.scatter(categories, values2D[:,1], c='b')
ax2.set_ylabel('wartości próbki')
ax2.set_xlabel('kategoria')
ax2.set_title('Przykładowe użycie funkcji scatter - dwie serie danych')
ax2.legend(['seria1', 'seria2'])
plt.show()
```

### Wykresy słupkowe - funkcja `bar`

```python
plt.bar(categories, values, color='r')
plt.ylabel('wartości próbki')
plt.xlabel('kategoria')
plt.title('Przykładowe użycie funkcji bar')
```

### Histogram

```python
x = np.random.poisson(5, 10000) #generujemy 10000 wartości z rozkładu Poissona o parametrze lambda równym 5
count, bins, ignored = plt.hist(x, 14, density=True)
plt.xlabel('Wartości')
plt.ylabel('Prawdopodobieństwo')
plt.title('Rozkład Poissona')
plt.text(10, .14, r'$\lambda$=5', fontdict={'size':16}) #współrzędne napisu, $ oznacza tryb pisania symboli matematycznych przy użyciu składni TeX
plt.grid(True)
plt.show()
```

### Zadanie 1

Wygeneruj:
    
1. 15000 wartości z rozkładu normalnego o $\mu = 14$, $\sigma = 5$
2. 15000 wartości z rozkładu normalnego o $\mu = 3$, $\sigma = 15$
3. 1500 wartości z rozkładu normalnego o $\mu = 10$, $\sigma = 7$


```python
dist1 = np.random.normal(14,5,15000)
dist2 = np.random.normal(3,15,15000)
dist3 = np.random.normal(10,7,1500)
```

Załóżmy, że wygenerowane dane to trzy serie pomiarów, a czas ich rejestracji jest taki sam (różnią się częstotliwością próbkowania). 

Narysuj wszystkie serie na jednym wykresie: 
- jedną oznacz tylko markerami (dowolnymi), 
- drugą linią kropka-kreska o zmienionej grubości, 
- trzecią linią z markerami.

Pamiętaj o opisach osi i legendzie. Wyświetl tylko wybrany fragment danych o długości 150 **próbek**.

```python
plt.figure()
plt.plot(dist1[0:150], 'xb')
plt.plot(dist2[0:150], '-.k',lw=1)
x = np.arange(0,150,10)
plt.plot(x,dist3[0:15], 'or-')
plt.show()
```
